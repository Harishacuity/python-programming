### Tried some image processing...

```python
import cv2 as cv
import numpy as np
path = r"/home/harish/Downloads/Black-dog-puppy.jpg"
img = cv.imread(path)
normalizedImg = np.zeros((800, 800))
normalizedImg = cv.normalize(img,  normalizedImg, 0, 150, cv.NORM_MINMAX)
cv.imshow('dst_rt', normalizedImg)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
from PIL import Image
from pylab import *

# read image to array
im = array(Image.open('/home/harish/Downloads/Black-dog-puppy.jpg'))

# plot the image
imshow(im)

# some points
x = [100,100,400,400]
y = [200,500,200,500]

# plot the points with red star-markers
plot(x,y,'r*')

# line plot connecting the first two points
plot(x[:2],y[:2])
# axis('off')

# add title and show the plot
title('Plotting: "Black-dog-puppy.jpg"')
show()
```
